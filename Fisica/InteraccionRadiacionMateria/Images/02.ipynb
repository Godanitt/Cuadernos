### Imagen 1: potencial $V_{FSN}$

```python
import matplotlib.pyplot as plt
import numpy as np 

def Vfsn(x):
    return (1/x)*(1-np.exp(-2*x))
x = np.linspace(0,10,100)
fig,ax=plt.subplots()
ax.plot(x,Vfsn(x),color="blue")
ax.set_xlabel("x/R")
ax.set_ylabel("$V_{FSN}$")
ax.grid(which='major', color='#DDDDDD', linewidth=0.8)
ax.grid(which='minor', color='#EEEEEE', linewidth=0.5)
ax.minorticks_on()
fig.savefig("02-Vfsn.pdf",bbox_inches="tight",dpi=300)
```
